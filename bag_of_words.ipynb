```python
import numpy as np
import pandas as pd
import csv
```

```python
lemmatized_train_df = pd.read_csv("data/lemmatized/train.csv")
lemmatized_test_df = pd.read_csv("data/lemmatized/test.csv")
```

```python
all_texts = lemmatized_train_df.text
```

```python
bow = dict()
```

```python
max_index = -1
for tweet_text in all_texts:
    for word in tweet_text.split(" "):
        word = word.lower()
        if word in bow or word == "":
            continue
        max_index+=1
        bow[word] = max_index
```

```python
for key, value in bow.items():
    bow[key] = value/max_index
```

```python
bow
```

```python
with open('data/lemmatized/bag_of_words/bow_test.csv', 'w') as f:
    for key, value in bow.items():
        f.write("".join([str(value),",\"",key, "\"\n"]))
```
